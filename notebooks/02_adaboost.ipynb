# AdaBoost Classifier — Concept & Implementation

**AdaBoost (Adaptive Boosting)** is a boosting ensemble technique that trains weak learners (typically shallow decision trees called *stumps*) **sequentially**.

### How it works:
1. Train a weak learner on the original data
2. Increase weights of **misclassified** samples
3. Train the next learner focusing on those hard samples
4. Repeat for `n_estimators` rounds
5. Final prediction = **weighted majority vote** of all weak learners

---

## 1. Imports

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier

print('✅ Libraries loaded')
```

## 2. Generate Synthetic Dataset

```python
X, y = make_classification(
    n_samples=500,
    n_features=20,
    n_informative=10,
    n_redundant=2,
    random_state=42
)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print(f'Train size: {X_train.shape[0]} | Test size: {X_test.shape[0]}')
```

## 3. Build AdaBoost Model

> **Base estimator**: Decision Tree Stump (`max_depth=1`) — the weakest possible learner

```python
# Decision Tree Stump as base weak learner
base_stump = DecisionTreeClassifier(max_depth=1)

# AdaBoost with 200 boosting rounds
ada_clf = AdaBoostClassifier(
    estimator=base_stump,
    n_estimators=200,
    random_state=42
)

ada_clf.fit(X_train, y_train)
print('✅ AdaBoost model trained')
```

## 4. Evaluation

```python
y_pred = ada_clf.predict(X_test)

print(f'Accuracy : {accuracy_score(y_test, y_pred):.4f}')
print(f'\nClassification Report:')
print(classification_report(y_test, y_pred))
```

## 5. Key Notes

| Parameter | Value | Reason |
|-----------|-------|--------|
| `estimator` | DecisionTreeClassifier(max_depth=1) | Weakest learner (stump) |
| `n_estimators` | 200 | More rounds = better correction |
| `learning_rate` | 1.0 (default) | Contribution of each tree |

**Advantage**: Works well even with weak learners, interpretable  
**Disadvantage**: Sensitive to noisy data and outliers (high weights on misclassified = overfit noise)
