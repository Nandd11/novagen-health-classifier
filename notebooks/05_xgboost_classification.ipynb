# XGBoost Classifier — Extreme Gradient Boosting

**XGBoost** is an optimized, high-performance implementation of Gradient Boosting. It adds several improvements over standard Gradient Boosting:

| Feature | Benefit |
|---------|--------|
| **Regularization (L1 + L2)** | Prevents overfitting |
| **Tree pruning (max_depth + gamma)** | More efficient tree growth |
| **Parallel processing** | Faster training |
| **Handling missing values** | Built-in sparse-aware split finding |
| **Column subsampling** | Like Random Forest — reduces overfitting |

> XGBoost consistently wins ML competitions (Kaggle) and is the industry standard for tabular data.

---

## 1. Installation & Imports

```python
# Uncomment if not installed
# !pip install xgboost

import xgboost as xgb
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

print(f'✅ XGBoost version: {xgb.__version__}')
```

## 2. Dataset

```python
X, y = make_classification(
    n_samples=500,
    n_features=20,
    n_informative=10,
    n_redundant=2,
    random_state=42
)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print(f'Train: {X_train.shape} | Test: {X_test.shape}')
```

## 3. Build & Train XGBoost Model

```python
xgb_clf = xgb.XGBClassifier(
    n_estimators=100,
    max_depth=3,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,   # column subsampling per tree
    eval_metric='logloss',
    use_label_encoder=False,
    random_state=42
)

xgb_clf.fit(
    X_train, y_train,
    eval_set=[(X_test, y_test)],
    verbose=False
)

print('✅ XGBoost model trained')
```

## 4. Evaluation

```python
y_pred = xgb_clf.predict(X_test)

print(f'Accuracy: {accuracy_score(y_test, y_pred):.4f}')
print('\nClassification Report:')
print(classification_report(y_test, y_pred))
```

## 5. XGBoost vs sklearn GradientBoosting

| Feature | sklearn GB | XGBoost |
|---------|-----------|--------|
| Speed | Slower | Much faster (parallel) |
| Regularization | No | Yes (L1/L2) |
| Missing values | Manual | Auto-handled |
| GPU support | No | Yes |
| Industry use | Academic | Production-grade |
